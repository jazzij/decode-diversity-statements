```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tag import pos_tag
from collections import Counter
from nltk.corpus import wordnet
from nltk.stem import PorterStemmer
import difflib
import pandas as pd
from nltk.stem import LancasterStemmer
lancaster=LancasterStemmer()
import os
from nltk.tag import pos_tag
porter = PorterStemmer()
import sys
import json 
```

```python
'''
Import all of the statements from the tokenText folder and extract the root and derivatives of root words
'''

statements = []

tokenizer = nltk.RegexpTokenizer(r"\w+")  # Removes all punctuation marks in the text
textdir = (r"C:\Users\manalais\decode-diversity-statements\tokenText")
for filename in os.listdir(textdir)[:]:
    filepath = os.path.join(textdir, filename)
    with open(filepath, "r") as file:
        text = file.read()
        tokens = tokenizer.tokenize(text)  # Returns a text as a list of words with punctuations
        tokens = [token.lower() for token in tokens if token.isalpha()]  # Changes all the words into lower case
        for word in tokens:
            statements.append(word)
    

#print(statements)
```

```python
'''
Import the uniquewordslist.txt which will be used for the root words
'''

textdir = (r'C:\Users\manalais\decode-diversity-statements')
filename = "uniquewordslist.txt"

text = []
with open(os.path.join(textdir, filename), 'r') as file:
        text = file.read()
        text = tokenizer.tokenize(text)
print(text)
```

```python
tagged_tokens = nltk.pos_tag(statements) # tag all the statements with the appropriate POS
```

```python
lemmatizer = WordNetLemmatizer()
dic = {}
```

```python
''' go through each word in tagged_tokens. Lemmatize earch word and add it to the dictionary with its leaves'''

for word, tag in tagged_tokens:
    if  not tag.startswith('NNP' or 'NNPS'): # remove propr nouns
        if tag.startswith('NN' or 'NNS'):
            lemmatized_word = lemmatizer.lemmatize(word, pos='n')
            if lemmatized_word in dic.keys():
                if word not in dic[lemmatized_word]:
                    dic[lemmatized_word].append(word)
            else:
                dic[lemmatized_word] = []
                dic[lemmatized_word].append(word)

    elif tag.startswith('VB' or 'VBP' or 'VBD' or 'VBG' or 'VBP' or 'VBZ'):
        lemmatized_word = lemmatizer.lemmatize(word, pos='v')
        if lemmatized_word in dic.keys():
            if word not in dic[lemmatized_word]:
                    dic[lemmatized_word].append(word)

        else:
            dic[lemmatized_word] = []
            dic[lemmatized_word].append(word)

    elif tag.startswith('JJ' or 'JJR' or 'JJS' ):
        lemmatized_word = lemmatizer.lemmatize(word, pos='a')
        if lemmatized_word in dic.keys():
            if word not in dic[lemmatized_word]:
                    dic[lemmatized_word].append(word)
        else:
            dic[lemmatized_word] = []
            dic[lemmatized_word].append(word)

    elif tag.startswith('RB' or 'RBR' or 'RBS' or 'RP'):
        lemmatized_word = lemmatizer.lemmatize(word, pos='r')
        if lemmatized_word in dic.keys():
            if word not in dic[lemmatized_word]:
                    dic[lemmatized_word].append(word)
        else:
            dic[lemmatized_word] = []
            dic[lemmatized_word].append(word)

    else:
        lemmatized_word = lemmatizer.lemmatize(word)
        if lemmatized_word in dic.keys():
            if word not in dic[lemmatized_word]:
                    dic[lemmatized_word].append(word)
        else:
            dic[lemmatized_word] = []
            dic[lemmatized_word].append(word)
```

```python
print(dic)
```

```python
'''create a list of keys that is not in the uniquewordlist.txt file'''
dic_keys = []
for key in dic.keys():
    if key not in text:
        dic_keys.append(key)
```

```python
''' remove all the keys that is not in uniquewordlist.txt file '''
for key in dic_keys:
    dic.pop(key)
```

```python
print(dic)
```

```python
'''Write dictionary to a text file'''
with open("stemnleaf.json", "w") as file:  
    json.dump(dic, file) 
```

```python
'''Visualize the data'''

key_list = []
value_list = []


for k in dic.keys():
    key_list.append(k)
for v in dic.values():
    value_list.append(v)
    
    
print("\n" + "\033[1m" + "Root and Leaf Plot".center(60) + "\n")
for key,value in zip(key_list,value_list):
    print("\033[1m" + key + "\033[0m" + '| ' + " ".join(value))
```
